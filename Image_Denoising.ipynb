# Image Denoising

All electronic devices are subject to noise -- pixels that, for one reason or another, take on an incorrect color or intensity. This means that the RGB values or light intensities do not accurately depict the event captured in the image.

*Noise* refers to unwanted or random fluctuations in the pixel values of an image. Noise can be caused by various factors such as the imaging sensor, electrical interference, or transmission errors.

## Smoothing Noisy Circle

### Smoothing With Mean

```python
Img = cv.imread('noisycircle.jpg', cv.IMREAD_GRAYSCALE);
cv2_imshow(Img)
```

```python
def smooth_mean(Img):
  Img_denoised = Img;
  row,col = Img.shape
  Img = Img.astype(np.float64)

  for i in range(1,row-1):
    for j in range(1,col-1):
      Img_denoised[i,j] = np.mean(Img[i-1:i+2,j-1:j+2]) # mean of neighboring cells

  return Img_denoised.astype(np.uint8)
```

```python
Img_denoised = smooth_mean(Img)
cv2_imshow(Img_denoised)
```

What happens if we iteratively denoise the image?

```python
Img = cv.imread('noisycircle.jpg', cv.IMREAD_GRAYSCALE);
cv2_imshow(Img)

for k in range(5):
  Img_denoised = smooth_mean(Img)
  Img = Img_denoised

cv2_imshow(Img_denoised)
```
