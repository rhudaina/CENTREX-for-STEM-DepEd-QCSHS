# Image Contrast

In an image, **contrast** is the measure of the difference in brightness between the light and dark areas of an image. In many cases, adjusting the contrast in an image can improve its quality. For example, in medical imaging, for example, this can help identify potential cancerous growths in X-rays of human tissue.

One way to increase contrast is by manipulating the intensity of the individual pixels in a digital image using transformations (functions).

```python
A = np.array([[17, 23, 47],
               [35,65,91],
               [67,89,100]])
print(A)
```

```python
plt.imshow(A, cmap = 'gray',vmin=0,vmax=255)
plt.axis("off")
plt.show()
```

## Linear Transformation

What happens if we perform a linear transformation
$$f(x) = 2x$$
on each pixel value $x$?

```python
Img = 2*A;
print(Img)

plt.imshow(Img, cmap = 'gray',vmin=0,vmax=255)
plt.axis("off")
plt.show()
```

## Quadratic Transformation

What happens if we perform a quadratic transformation
$$f(x) = x^2$$
on each pixel value $x$?

```python
Img = np.square(A);
print(Img)

plt.imshow(Img, cmap = 'gray',vmin=0,vmax=255)
plt.axis("off")
plt.show()
```

### Normalized Quadratic Transformation

After quadratic transformation, normalize the matrix by dividing each pixel by
$$\frac{1}{255}\times\text{maximum pixel intensity}$$

```python
Img = np.square(A);

norm = np.max(Img)/255;
print(norm)

Img = Img/norm;
print(Img)
```

```python
plt.imshow(Img, cmap = 'gray',vmin=0,vmax=255)
plt.axis("off")
plt.show()
```
